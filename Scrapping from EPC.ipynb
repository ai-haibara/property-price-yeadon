```python
def main():
    # create lists to store our data
    all_surface = []
    
    total_number_of_properties = 850 # 425*2
    
    v=0
    
    while v < total_number_of_properties:
    #for v in [4]:
    #for postcode in list(POSTCODE.keys()):
            print(properties_list[v])
            print(properties_list[v+1])
            # define our user headers
            headers = {
                "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.114 Safari/537.36"
            }

            # the website displays a list of certificates associated to the postcode (each property has its own certification id)
            find_energy_certificate = f"https://find-energy-certificate.service.gov.uk/find-a-certificate/search-by-postcode?postcode={properties_list[v]}"
            
            #Wait for 0.3 seconds to prevent the targetted website to crash
            time.sleep(0.3)
            
            # request our webpage
            res = requests.get(find_energy_certificate, headers=headers)

            # check status
            res.raise_for_status()

            soup = BeautifulSoup(res.text, "html.parser")
            
            # This gets the list of properties
            list_property = soup.find_all("tbody", class_="govuk-table__body")
            
           # list of certificates and their associated addresses in separate lists
            all_url = []
            all_addresses = []
            
            for link in soup.find_all('a', attrs={'href': re.compile("/energy-certificate/")}): 
                # get the urls and the addresses displayed on the page in capital letters for better comparison
                all_url.append(link.get('href'))
                all_addresses.append(link.get_text().strip().upper())
                
            print('all url')
            print (all_url)
            print('all addresses')
            print (all_addresses)
            print(len(all_url))
            
            # Get the property street number or name
            check_address = str(properties_list[v+1])
            print('property number or name')
            print(check_address)
            index = 0
            i = 0
            j = 0
            print('length of all addresses')
            print(len(all_addresses))
            
            #Copy the first n elements of all_addresses into a new list with n being the length of characters of check address 
            print('length check address')
            print(len(str(check_address)))
            all_addresses_street_number = []
            for i in range(len(all_addresses)):
                list_element = all_addresses[i]
                temp_list=[]
                temp_list.extend(list_element)
                print('temp list')
                print(temp_list)
    
                temp_list_element = []
                for j in range(len(str(check_address))):    
                    #print(temp_list[j])
                    temp_list_element.append(temp_list[j])
                    element = "".join(temp_list_element)
                    #print('element')
                    #print(element)
                all_addresses_street_number.append(element)
            print('all addresses street number')
            print(all_addresses_street_number)
            
            # Compare the property name/ street number with the list of all_addresses to get the associated url link
            
            #case where the property epc is not registered
            if len(all_addresses_street_number)==0:
                floor_area = 'unknown information'
            else:
                #special case len(check_address)= 1
                #check for index = 0
                if len(str(check_address))==1: 
                    if str(check_address) == str(all_addresses_street_number[0]):                       
                        #get the index of the list of all_addresses to get the associated url
                        index = 0 
                        floor_area = ''
                        print('index')
                        print(index)
                        print('all url index')
                        print(all_url[index])
                    else:
                        i=1
                        #for i in range(len(all_addresses_street_number)):
                        while ((index == 0) and (i < (len(all_addresses_street_number)))):
                            print('loop addresses')
                            print(index)
                            if str(check_address) == str(all_addresses_street_number[i]):                       
                                #get the index of the list of all_addresses to get the associated url
                                index = i 
                                floor_area = ''
                                print('index')
                                print(index)
                                print('all url index')
                                print(all_url[index])
                    
                            else:
                                print('index different from 0 and comparison not valid')
                                i+=1
                                floor_area = 'error'
                                
                elif str(check_address) == str(all_addresses_street_number[0]):
                    print('length of check address different from 0 but check address equals first element')
                    #get the index of the list of all_addresses to get the associated url
                    index = 0 
                    floor_area = ''
                    print('index')
                    print(index)
                    print('all url index')
                    print(all_url[index])
                
                else:
                    i=0
                    #for i in range(len(all_addresses_street_number)):
                    while ((index == 0) and (i < (len(all_addresses_street_number)))):
                        print('loop addresses')
                        print(index)
                        if str(check_address) == str(all_addresses_street_number[i]):                       
                            #get the index of the list of all_addresses to get the associated url
                            index = i 
                            floor_area = ''
                            print('index')
                            print(index)
                            print('all url index')
                            print(all_url[index])
                            
                    
                        else:
                            print('index different from 0 and comparison not valid')
                            i+=1
                            floor_area = 'error'     
            print('next step')
            print(floor_area)
            print(index)
            
            if floor_area == '':
                #go to the associated url
                find_house_surface = f"https://find-energy-certificate.service.gov.uk{all_url[index]}"
                    
                # Wait for 1 second to prevent the website to crash
                time.sleep(1)
                    
                # request our webpage
                request_epc = requests.get(find_house_surface, headers=headers)

                # check status
                request_epc.raise_for_status()

                epc_soup = BeautifulSoup(request_epc.text, "html.parser")
            
                property_information = []
            
                for information in epc_soup.find_all("dd", class_="govuk-summary-list__value govuk-!-width-one-half"): 
                # find Total floor area
                    property_information.append(information.get_text().strip())
                    #print(property_information)
                
                #print(property_information)
                floor_area = property_information[1]
                print('floor_area')
                print(floor_area)        
                    
                all_surface.append(floor_area)
                
            else:
                #floor area would be an error
                all_surface.append(floor_area)
            #print(all_surface)      
            v+=2
            
    # create list to store all the data
    data = [] 
            
    h=0
    while h < total_number_of_properties/2:
        inside_list = []
        inside_list.append(postcode_list[h])
        inside_list.append(address_list[h])
        inside_list.append(all_surface[h])
        
        print(inside_list)
        
        data.append(inside_list)
        
        h+=1
        
        
        
    # convert data to dataframe    
    #print(data)
    
    df = pd.DataFrame(data, columns =['Postcode', 'Address number', 'Surface area'], dtype = float) 
    
    #print(df)
    
    #convert dataframe to csv file
    df.to_csv('Surface_data4.csv', index=False)
            
if __name__ == "__main__":
    main()
```
